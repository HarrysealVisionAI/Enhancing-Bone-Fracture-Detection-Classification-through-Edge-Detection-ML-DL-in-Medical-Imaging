```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the image
models = ['Deep ConvNet', 'ResNet', 'VGG16-SVM', 'Proposed Deep ConvNet + Canny']
fracture = [0.98, 0.97, 0.84, 1]
healthy = [0.89, 0.96, 0.83, 0.96]

# Bar width and separation
bar_width = 0.35
bar_spacing = 0.2 
index = np.arange(len(models))

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(index - (bar_width/2 + bar_spacing/2), fracture, bar_width, label='Fracture', color='red')
bars2 = ax.bar(index + (bar_width/2 + bar_spacing/2), healthy, bar_width, label='Healthy', color='green')

# Adding labels and title
ax.set_xlabel('Models')
ax.set_ylabel('Values')

# Set title with increased font size
ax.set_title('Precision comparison of the advanced method with hybrid deep ConvNet integrated canny edge detection', fontsize=12.5)
ax.set_xticks(index)
ax.set_xticklabels(models, fontweight='bold')

# Adding the percentage labels above bars
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', ha='center', va='bottom')

# Adding Fracture and Healthy labels below the bars and making them bold
for i in range(len(models)):
    ax.text(index[i] - (bar_width/2 + bar_spacing/2), -0.05, 'Fracture', ha='center', va='top', color='red', fontsize=10, weight='bold')
    ax.text(index[i] + (bar_width/2 + bar_spacing/2), -0.05, 'Healthy', ha='center', va='top', color='green', fontsize=10, weight='bold')


ax.legend()
plt.tight_layout()
plt.show()
```
